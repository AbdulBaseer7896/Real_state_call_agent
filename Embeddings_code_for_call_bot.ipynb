```python
import json
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.schema import Document
import pandas as pd
```

```python
def create_vectorstore():
    file_path = "ar_split.json"

    # Load merged JSON (a list of dicts)
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    docs = []


    for item in data:
      doc = Document(
          page_content=str(item),
      )
      docs.append(doc)

    print(f"Created {len(docs)} chunks with metadata.")

    # Define the embedding model
    model_name = "BAAI/bge-m3"
    model_kwargs = {"device": "cuda"}
    encode_kwargs = {"normalize_embeddings": True, "batch_size": 8}

    embedding_model = HuggingFaceEmbeddings(
        model_name=model_name,
        model_kwargs=model_kwargs,
        encode_kwargs=encode_kwargs
    )

    vectorstore = FAISS.from_documents(docs, embedding_model)
    vectorstore.save_local("./faiss_index")
    print("Vectorstore saved to ./faiss_index")
```

**For .xlsx files**

```python
def create_vectorstore():
    file_path = "Birthday_party_packages.xlsx"  # Change to your actual CSV filename

    # Load CSV file using pandas
    df = pd.read_excel(file_path)

    docs = []

    # Convert each row into a Document (you can customize which columns to include)
    for _, row in df.iterrows():
        content = " ".join(str(row[col]) for col in df.columns)  # or specify columns like: [row['title'], row['body']]
        doc = Document(page_content=content)
        docs.append(doc)

    print(f"Created {len(docs)} documents from excel file.")

    # Define the embedding model
    model_name = "BAAI/bge-m3"
    model_kwargs = {"device": "cuda"}
    encode_kwargs = {"normalize_embeddings": True, "batch_size": 8}

    embedding_model = HuggingFaceEmbeddings(
        model_name=model_name,
        model_kwargs=model_kwargs,
        encode_kwargs=encode_kwargs
    )

    vectorstore = FAISS.from_documents(docs, embedding_model)
    vectorstore.save_local("./faiss_index_birthday")
    print("Vectorstore saved to faiss_index folder")
```

```python
create_vectorstore()
```
